```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Math": [78, 85, 62, 90, 55, 88, 70, 95],
    "Science": [80, 82, 60, 92, 58, 85, 75, 94],
    "English": [75, 88, 65, 91, 60, 90, 72, 93],
    "StudyHours": [2, 3, 1, 4, 1, 3, 2, 5]
}

df = pd.DataFrame(data)
df
```

```python
df['TotalScore']=df['Math']+df['Science']+df['English']
df
```

```python
df[df['TotalScore']==df['TotalScore'].max()]['Name']
```

```python
import matplotlib.pyplot as plt
x=df['Name']
y=df['TotalScore']
plt.figure(figsize=(8,5))
plt.bar(x,y,color='skyblue',edgecolor='Black')
plt.xlabel('Name')
plt.ylabel('TotalScore')
plt.title('student marks analysis')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python

x=df['StudyHours']
plt.figure(figsize=(8,5))
plt.hist(x,color='skyblue',edgecolor='black',bins=5)
plt.ylabel('Frequency')
plt.xlabel('StudyHours')
plt.title('student marks analysis')

plt.tight_layout()
```

```python
plt.figure(figsize=(8,5))
x=df['Math']
y=df['Science']
z=df['English']
box=plt.boxplot([x,y,z],patch_artist=True)
plt.title('student marks analysis')
plt.ylabel('subjects')
plt.xlabel('marks')
plt.xticks([1,2,3],['Math','Science','English'])
colors = ['skyblue', 'lightgreen', 'salmon']
for patch,color in zip(box['boxes'],colors):
  patch.set_facecolor(color)
```
